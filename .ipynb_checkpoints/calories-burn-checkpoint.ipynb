```python
X = df.drop(['Calories', 'User_ID'], axis="columns")
y = df.Calories
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

#### Decision Tree

```python
step1 = ColumnTransformer(transformers=[('col_tnf', OneHotEncoder(sparse=False, drop='first'), ['Gender'])], remainder='passthrough')

step2 = DecisionTreeRegressor()
```

```python
pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
pipe.score(X_test, y_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
print('R2 score', r2_score(y_test,y_pred))
print('MAE', mean_absolute_error(y_test,y_pred))
```

#### Random Forest Regressor

```python
rfstep1 = ColumnTransformer(transformers=[('col_tnf', OneHotEncoder(sparse=False, drop='first'), ['Gender'])], remainder='passthrough')

rfstep2 = RandomForestRegressor()

rfpipe = Pipeline([
    ('rfstep1', rfstep1),
    ('rfstep2', rfstep2)
])
rfpipe.fit(X_train, y_train)

y_predrf = pipe.predict(X_test)

rfpipe.score(X_test, y_test)
```

```python
print('R2 score', r2_score(y_test,y_predrf))
print('MAE', mean_absolute_error(y_test,y_predrf))
```
